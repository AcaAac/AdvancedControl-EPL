```python
import pandas as pd

time_data = pd.read_csv('time_data.csv')
lyapunov_data = pd.read_csv('lyapunov_data.csv')
dv_data = pd.read_csv('dv_data.csv')
```

```python
import matplotlib.pyplot as plt

# Assuming the data you want to plot is in the first column
time = time_data.iloc[:, 0]  # Adjust if your time data is in a different column
lyapunov_values = lyapunov_data.iloc[:, 0]  # Adjust similarly
dv_values = dv_data.iloc[:, 0]  # Adjust similarly
plt.plot(time, lyapunov_values)
plt.xlabel('Time [s]')
plt.ylabel('Lyapunov Function [a.u.]')
plt.title('Lyapunov Data over Time')
plt.show()
plt.plot(time, dv_values)
plt.xlabel('Time [s]')
plt.ylabel('Lyapunov Derivative [a.u.]')
plt.title('Lyapunov Derivative Data over Time')
plt.show()
```
